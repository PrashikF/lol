### Reshaping in Pandas

```python
# | Method              | Use Case                         |
# | ------------------- | -------------------------------- |
# | pivot()             | Long ➜ Wide (no duplicates)      |
# | pivot_table()       | Long ➜ Wide (duplicates allowed) |
# | melt()              | Wide ➜ Long                      |
# | stack()             | Columns ➜ Rows                   |
# | unstack()           | Rows ➜ Columns                   |
# | T / transpose()     | Swap rows & columns              |
```

```python
# pivot() – Long ➜ Wide format
# Used when data has unique index–column pairs.

df = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02'],
    'City': ['Pune', 'Mumbai', 'Pune'],
    'Temp': [30, 32, 28]
})

df_pivot = df.pivot(index='Date', columns='City', values='Temp')
print(df_pivot)
```

```python
# pivot_table() – handles duplicate values
# More powerful than pivot().
df.pivot_table(
    index='Date',
    columns='City',
    values='Temp',
    aggfunc='mean'
)
```

```python
# melt() – Wide ➜ Long format
# Opposite of pivot().

df = pd.DataFrame({
    'City': ['Pune', 'Mumbai'],
    'Jan': [30, 32],
    'Feb': [28, 31]
})

df_melt = pd.melt(
    df,
    id_vars=['City'],
    value_vars=['Jan', 'Feb'],
    var_name='Month',
    value_name='Temperature'
)

print(df_melt)
```

```python
# stack() – Columns ➜ Rows
# Converts columns into row index.
# unstack() – Rows ➜ Columns
# Reverse of stack().

df = pd.DataFrame({
    'Math': [80, 85],
    'Science': [90, 88]
}, index=['A', 'B'])

print(df)
print("===========================")
df_stack = df.stack()
print(df_stack)
print("===========================")
df_unstack = df_stack.unstack()
print(df_unstack)
```

```python
# transpose() / T – swap rows & columns
# Very simple reshaping.
df.T
```
